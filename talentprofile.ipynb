```python
import nltk

nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python
import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

stop_words = ENGLISH_STOP_WORDS

def clean_text(text):
    # Convert to string
    text = str(text).lower()

    # Remove HTML
    text = re.sub(r"<.*?>", " ", text)

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", " ", text)

    # Keep only letters
    text = re.sub(r"[^a-zA-Z\s]", " ", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Remove stopwords
    words = [
        word for word in text.split()
        if word not in stop_words
    ]

    return " ".join(words)
```

```python
talent_profiles["clean_text"] = talent_profiles["combined_text"].apply(clean_text)
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.metrics.pairwise import cosine_similarity

import numpy as np
```

```python
vectorizer = TfidfVectorizer(
    max_features=10000,
    ngram_range=(1,2)
)

tfidf_matrix = vectorizer.fit_transform(
    talent_profiles["clean_text"]
)

print(tfidf_matrix.shape)
```

```python
def recommend_talent(query, top_n=10):

    # Clean query
    query = clean_text(query)

    # Convert query into TF-IDF vector
    query_vector = vectorizer.transform([query])

    # Calculate similarity
    similarity = cosine_similarity(
        query_vector,
        tfidf_matrix
    ).flatten()

    # Top matching indices
    top_indices = similarity.argsort()[-top_n:][::-1]

    # Create result dataframe
    results = talent_profiles.iloc[top_indices].copy()

    results["Similarity Score"] = similarity[top_indices]

    return results[
        [
            "title",
            "company_name",
            "industry_name",
            "skill_name",
            "formatted_experience_level",
            "location",
            "Similarity Score"
        ]
    ]
```

```python
recommend_talent(
    "Need a motion graphics designer for real estate"
)
```

```python
recommend_talent(
    "Need a UI UX designer"
)
```

```python
def display_recommendations(query):

    results = recommend_talent(query)

    print("=" * 80)
    print(f"Query: {query}")
    print("=" * 80)

    for idx, (_, row) in enumerate(results.iterrows(), start=1):

        print(f"\n🏆 Rank {idx}")

        print(f"Title : {row['title']}")

        print(f"Company : {row['company_name']}")

        print(f"Industry : {row['industry_name']}")

        print(f"Skills : {row['skill_name']}")

        print(f"Experience : {row['formatted_experience_level']}")

        print(f"Location : {row['location']}")

        print(f"Similarity : {row['Similarity Score']:.3f}")
```

```python
display_recommendations(
    "Need someone who creates luxury real estate reels using motion graphics"
)
```

```python
display_recommendations("Need a Python Data Analyst")

display_recommendations("Need a UI UX Designer")

display_recommendations("Need a Motion Graphics Designer")

display_recommendations("Need a Real Estate Marketing Expert")

display_recommendations("Need a Machine Learning Engineer")
```
